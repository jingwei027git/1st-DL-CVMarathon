## 1. 取出今天所有的新聞

```python
import requests
import datetime
import time
from bs4 import BeautifulSoup
from selenium import webdriver

articles = []
article_last = ''

def load_continue():
    for i in range(3):
        time.sleep(2)
        browser.execute_script("window.scrollTo(0, 900000);")
        print('refresh...')

def load_articles():
    global articles
    global article_last
    articles = []
    soup = BeautifulSoup(browser.page_source, "html5lib")
    for d in soup.find(class_="part_list_2").find_all('h3'):
        article = d.find(class_="date").text + ' - ' + d.find_all('a')[-1].text
        article_last = article
        articles.append(article)
        print('[', article, ']')

today = datetime.datetime.now().strftime("%Y/%m/%d")
print(today)

browser = webdriver.Chrome(executable_path='/Users/david/Projects/python-marathon/chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
load_continue()
load_articles()
print(article_last)
while article_last.startswith(today) == True:
    load_continue()
    load_articles()

articles_text = '\n'.join(map(str, articles))
```
