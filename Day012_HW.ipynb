## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
# 0. 讀數據
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

### ① 取出今天所有的發文

```python
# 今天所有發文
articles = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    articles.append(d.find_all('a')[-1].text)

print(articles)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# [2] 用 dict 來放
category_map = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    category = d.find('em').text
    if not(category in category_map.keys()):
        category_map[category] = []
    category_map.get(category).append({
        'date': d.find(class_="date").text,
        'title': d.find_all('a')[-1].text
        })
print(category_map)
```

### ③ 哪一個類別的文章最多

```python
max_count = -1
max_category = ''
for k in category_map.keys():
    count = len(category_map.get(k))
    if (count > max_count):
        max_count = count
        max_category = category

print(max_category, max_count)
```
